```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob

%matplotlib inline
```

```python
files = glob.glob('*.csv')
wins = []
for name in files:
    print(name)
    df = pd.read_csv(name, skipinitialspace=True, quotechar="'")
    x = df.groupby('seed')['outcome'].apply(lambda x: x.to_list()[0]).value_counts()     
    wins.append(x['win'] / x.sum())
```

```python
y_pos = np.arange(len(files))
plt.figure(figsize=(18, 10))
plt.barh(y_pos, wins, align='center', alpha=0.5)
plt.yticks(y_pos, files)
plt.xticks(np.arange(0, 1.1, .25))
plt.title('% Wins')
plt.show()
print(list(zip(files, wins)))
```

```python
won = []
loss = []
for name in files:
    df = pd.read_csv(name)
    x = df.groupby('outcome')['rounds']
    x = x.sum() / x.count()
    won.append(x['win'])
    loss.append(x['loss'])
```

```python
plt.figure(figsize=(18, 5))
plt.barh(y_pos, won, align='center', alpha=0.5)
plt.yticks(y_pos, files)
plt.xlim(30, 45)
plt.title('Number of rounds of Won games')
plt.show()
print(list(zip(files, won)))
```

```python
plt.figure(figsize=(18, 5))
plt.barh(y_pos, loss, align='center', alpha=0.5)
plt.yticks(y_pos, files)
plt.xlim(30, 45)
plt.title('Num rounds of lost games')
plt.show()
print(list(zip(files, loss)))
```
